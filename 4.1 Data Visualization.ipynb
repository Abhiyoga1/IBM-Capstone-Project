## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

QUERY = """ SELECT ConvertedComp FROM master """

df_Con = pd.read_sql_query(QUERY, conn)

sns.histplot(df_Con, bins=60, color='skyblue', edgecolor='black')
plt.title('Histogram ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """ SELECT Age FROM master"""

df_age = pd.read_sql_query(QUERY, conn)

sns.boxplot(data=df_age)
plt.title('Boxplot Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """ SELECT Age, WorkWeekHrs FROM master"""

df_scat = pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=df_scat, x='Age', y='WorkWeekHrs')
plt.title('Scatter Plot WorkWeekHrs Vs Age')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here



QUERY = """ SELECT Age,WorkWeekHrs,CodeRevHrs FROM master"""

df_bub = pd.read_sql_query(QUERY, conn)

plt.scatter(df_bub['WorkWeekHrs'], df_bub['CodeRevHrs'], s=df_bub['Age'])
plt.title('Buble Plot WorkWeekHrs VS CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY  = """ SELECT Respondent,DatabaseDesireNextYear FROM DatabaseDesireNextYear"""

df_pie = pd.read_sql_query(QUERY,conn)

df_pie=pd.DataFrame(df_pie['DatabaseDesireNextYear'].value_counts())
df_pie = df_pie.head(5)
fig = px.pie(df_pie, values= ((df_pie.values/ df_pie.values.sum())*100).flatten(), names=df_pie.index, title= "Percentage Database Wish To Learn")
fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """ SELECT WorkWeekHrs,CodeRevHrs,Age FROM master WHERE Age >=30 AND Age <=35"""

df_stack = pd.read_sql_query(QUERY,conn)
df_stack= df_stack.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

x = df_stack.index

plt.bar(x , df_stack['WorkWeekHrs'], color = 'b')
plt.bar(x ,  df_stack['CodeRevHrs'], bottom=df_stack['WorkWeekHrs'], color = 'r'  )
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend(['WorkWeekHrs', 'CodeRevHrs'])
plt.title('Stacked Chrat of Median WorkWeekHrs and CodeRevHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """ SELECT ConvertedComp,Age FROM master WHERE Age >=45 AND Age <=60"""

df_line = pd.read_sql_query(QUERY,conn)

df_line = df_line.groupby('Age')['ConvertedComp'].median()

plt.plot(df_line.index, df_line.values)
plt.title('Line Chart of ConvertedComp Median From Age 45 - 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """ SELECT MainBranch FROM master"""

df_bar  =pd.read_sql_query(QUERY, conn)
df_bar= df_bar.value_counts()

df_bar = df_bar.reset_index()
plt.barh(df_bar['MainBranch'], df_bar['count'])
plt.title('Bar Chart of Mainbranch')
plt.xlabel('count')
plt.show()
```
